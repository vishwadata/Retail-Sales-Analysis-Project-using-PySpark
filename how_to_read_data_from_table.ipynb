```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("MyFirstPySpark") \
    .getOrCreate()

table_df=spark.read.table("samples.bakehouse.sales_transactions")
table_df.show()
table_df.count()

from pyspark.sql.functions import concat_ws,col
table_df.withColumn("product_payment",concat_ws("-",col("product"),col("paymentMethod"))).show(10,False)


```
